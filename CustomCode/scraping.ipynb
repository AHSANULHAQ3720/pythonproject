```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://bahriatown.com/index.php"
data = requests.get(url)
html = data.text

soup = BeautifulSoup(html, 'html5lib')

print(soup.prettify())

# Title of Source Page
print(soup.title)
```

```python
# Search everything with bold tag
print(soup.h3)
print(soup.h3.next_siblings)
```

```python
# Find All Link on index page
for link in soup.find_all("a",href=True):
    print(link.get('href'))
```

```python
# Scrape all images
for img in soup.find_all("img"):
    # print(img)
    print(img.get("src"))
```
